```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

## Read Data

```python
data = pd.read_csv("nba.csv", index_col ="Name") 
```

```python
data.head()
```

```python
data.info()
```

```python
data = data.dropna()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

## Bar Histogram

```python
plt.hist(data['Age'])
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.hist(data['Age'])
plt.show()
```

```python
plt.figure(figsize=(16,5))

data.Age.value_counts().plot(kind='bar')
plt.show()
```

You can plot with dataframe different kinds of plots as:
* ‘bar’ or ‘barh’ for bar plots
* ‘hist’ for histogram
* ‘box’ for boxplot
* ‘kde’ or ‘density’ for density plots
* ‘area’ for area plots
* ‘scatter’ for scatter plots
* ‘hexbin’ for hexagonal bin plots
* ‘pie’ for pie plots

```python
data.Age.plot(kind='kde')
plt.show()
```

```python
data.Age.plot.kde()
```

```python
ax = data[['Age', 'Number']][:100].plot.area()
```

## Line Histogram

```python
plt.figure(figsize = (10,6))
plt.plot(data['Age'][:5])
plt.show()
```

## Combination bar and line plot

```python
##Combination
import math
columnNumber = 2
data_columns = data._get_numeric_data().columns.values
plot_columns = columnNumber
number_of_columns = math.ceil(len(data_columns)/plot_columns)
number_of_rows = math.ceil((len(data_columns))/number_of_columns)

plt.figure(figsize=(12, 12))
    
for i in range(0, len(data_columns)):
    plt.subplot(number_of_columns + 1, number_of_rows, i+1)
    sns.distplot(data[data_columns[i]], kde=True) 
```

## Scatter plot

```python
plt.figure(figsize = (10,6))
plt.scatter(x=data['Age'], y=data['Salary'], alpha=0.3)

plt.ylabel('Salary')
plt.xlabel('Age')
plt.show()
```

## Box plots

```python
plt.boxplot(data['Age'], labels=['Age'])
plt.show()
```

```python
data.groupby(['Team'])['Age'].plot(kind='box')
plt.show()
```

## Comparison plots

```python
data_columns = data.columns.values
data_columns = data._get_numeric_data().columns.values
plot_columns = 4
number_of_columns = math.ceil(len(data_columns)/plot_columns)
number_of_rows = math.ceil((len(data_columns))/number_of_columns)

plt.figure(figsize=(15,8))
for i in range(0,len(data_columns)):
    plt.subplot(number_of_columns, number_of_rows, i+1)
    sns.set_style('whitegrid')
    sns.boxplot(data[data_columns[i]],palette='viridis',orient='v')
    plt.tight_layout()
```

```python
plt.figure(figsize=(16,5))

data.Team.value_counts().plot(kind='bar')
plt.show()
```

```python
plt.figure(figsize=(16,15))

data.Team.value_counts().plot(kind='barh')
plt.show()
```

```python
plt.figure(figsize=(15, 20))
sns.catplot(x="Height", y="Weight", hue="Age", kind="swarm", data=data, order = 
            ['5-9', '6-0','6-1', '6-2','6-3','6-4','6-5','6-6','6-7','6-8','6-9','7-0','7-1','7-2']);
```

```python
data.groupby('Position').mean()
```

```python
data.groupby('Position').mean()['Weight'].plot(kind='line')
plt.show()
```

## Composition plots

```python
# Prepare Data
df = data.groupby('Height').size()

# Make the plot with pandas
df.plot(kind='pie', subplots=True, figsize=(8, 8))
plt.title("Pie Chart of Heights")
plt.ylabel("")
plt.show()
```

## Relationship plots

```python
correlation = data._get_numeric_data().corr()
sns.heatmap(correlation, annot=True, cmap = 'viridis')
```
